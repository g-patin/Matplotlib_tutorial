# Create multiple subplots 

## Simple subplots

In this section, the subplots have all the same dimensions and the space (width and height) between the subplots is equally distributed.

If you need to create more complex subplots, look at the following sections : 'Complex subplots'

### Example 1 

```python
# retrieve the data to be plottted
x = df['t_sec']
y1 = df['dE76']
y2 = df['dE00']
y3 = df['a*']
y4 = df['L*']

# create a figure and 4 axes in a 1x4 combination
fig, [ax1, ax2, ax3, ax4] = plt.subplots(1,4, figsize=(20,5))

# plot the data inside inside the axis
ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
ax4.plot(x,y4)

# show the figure
plt.show()
```

### Example 2 

```python
# retrieve the data to be plottted
x = df['t_sec']
y1 = df['dE76']
y2 = df['dE00']
y3 = df['a*']
y4 = df['L*']

# create a figure and 4 axes in a 2x2 combination
fig, [[ax1, ax2],[ax3, ax4]] = plt.subplots(2,2, figsize=(15,7))

# plot the data inside inside the axis
ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
ax4.plot(x,y4)

# show the figure
plt.show()
```

## Complex subplots 

There are several ways to create precisely adjust the layout of subplots. Here we are showing one of the easiest way, which is to define additional parameters in the `plt.subplots()` function. 

### Example 1 - adjust subplots size

In this example, we adjust the size of each subplot by defining two parameters (**width_ratios** and **height_ratios**) inside the `plt.subplots()` function.

To avoid errors when defining these two parameters, it is important to be aware that the amount of values given inside each parameter should be equal to amount of rows and columns. In this example, I have 1 row and 4 columns, therefore the parameter 'width_ratios' must contain 4 numerical values. 

Here, I defined `width_ratios=(3,3,3,1.5)`, which means that last subplot is two times smaller than the others, I could have also written `width_ratios=(2,2,2,1)` and obtain the same output.


```python
# retrieve the data to be plottted
x = df['t_sec']
y1 = df['dE76']
y2 = df['dE00']
y3 = df['a*']
y4 = df['L*']

# create a figure and 4 axes in a 1x4 combination
fig, [ax1, ax2, ax3, ax4] = plt.subplots(1,4, figsize=(20,5), width_ratios=(3,3,3,1.5))

# plot the data inside inside the axis
ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
ax4.plot(x,y4)

# show the figure
plt.show()
```

### Example 2 - adjust subplots size 

Same as example 1, but with more subplots. Here the first column is twice bigger than the second one and the first row is three times smaller than the other rows.

```python
# retrieve the data to be plottted
x = df['t_sec']
y1 = df['dE76']
y2 = df['dE00']
y3 = df['a*']
y4 = df['L*']
y5 = df['b*']
y6 = df['C*']

# create a figure and 6 axes in a 3x2 combination
fig, [[ax1, ax2],[ax3, ax4], [ax5, ax6]] = plt.subplots(3,2, figsize=(15,7), width_ratios=(2,1), height_ratios=(0.1,0.3,0.3))

# plot the data inside inside the axis
ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
ax4.plot(x,y4)
ax5.plot(x,y5)
ax6.plot(x,y6)

# show the figure
plt.show()
```

### Example 3 - adjust subplots spacing

To adjust the spaces between the subplots, you will need to add a new function (`plt.subplots_adjust()`) inside which you will define the height space (**hspace**) and the width space (**wspace**). The space will be equally distributed if you have more than 2 rows or columns. If you want uneven space distribution, you can do that with the `GridSpec()`, but it is a bit more complicated.

In addition you can also adjust the space in the margins of the subplos (**top**, **bottom**, **left**, **right**).

```python
# retrieve the data to be plottted
x = df['t_sec']
y1 = df['dE76']
y2 = df['dE00']
y3 = df['a*']
y4 = df['L*']
y5 = df['b*']
y6 = df['C*']

# create a figure and 6 axes in a 3x2 combination
fig, [[ax1, ax2],[ax3, ax4], [ax5, ax6]] = plt.subplots(3,2, figsize=(15,7), width_ratios=(2,1), height_ratios=(0.1,0.2,0.3))

# adjust the subplots space
plt.subplots_adjust(wspace=0.5, hspace=0.3)  # if you want, you can add: left=0.125, right=0.9, top=1.5, bottom=0.1

# plot the data inside inside the axis
ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
ax4.plot(x,y4)
ax5.plot(x,y5)
ax6.plot(x,y6)




# show the figure
plt.show()
```

# Subplots with shared axes 

## Example 1 

```python
# retrieve the data to be plottted
x = df['t_sec']
y1 = df['dE76']
y2 = df['dE00']

fig, axes = plt.subplots(2,1, sharex= True)
axes[0].plot(x,y1)
axes[1].plot(x,y2)

plt.tight_layout()

plt.show()
```

## Example 2 

```python
# retrieve the data to be plottted
x = df['t_sec']
y1 = df['dE76']
y2 = df['dE00']

fig, axes = plt.subplots(1,2, sharey= True) # synchronisation des axes en mode interactif
axes[0].plot(x,y1)
axes[1].plot(x,y2)

plt.tight_layout()

plt.show()
```

## Example 3 

```python
# retrieve the data to be plottted
x = df['t_sec']
y1 = df['dE76']
y2 = df['dE00']
y3 = df['a*']
y4 = df['L*']

fig, axes = plt.subplots(2,2, sharex=True, sharey=True)
axes[0,0].plot(x,y1, label='top left')
axes[0,1].plot(x,y2, label='top right')
axes[1,0].plot(x,y3, label='bottom left')
axes[1,1].plot(x,y4, label='bottom right')

axes[0,0].legend()
axes[0,1].legend()
axes[1,0].legend()
axes[1,1].legend()

plt.tight_layout()

plt.show()
```

# Add axes labels 

```python
# retrieve the data to be plottted
x = df['t_sec']
y = df['L*']


# create the figure and the axis
fig, ax = plt.subplots()

# plot the data
ax.plot(x,y)

# add x and y labels
ax.set_xlabel('x label')
ax.set_ylabel(r'$\bf{y lab}$el', fontsize = 15, color = 'blue')   # set the fontsize to 15, the color to blue, and make it partially bold

# show the plot
plt.show()
```

# Add title

```python
# retrieve the data to be plottted
x = df['t_sec']
y = df['L*']

# create the figure and the axis
fig, ax = plt.subplots()

# plot the data
ax.plot(x,y)

# add a title
ax.set_title(r'A title with greek letter $\Delta$', fontsize=15, color='red')

# show the plot
plt.show()
```

# Add legend 

```python
# retrieve the data to be plottted
x = df['t_sec']
y1 = df['dE76']
y2 = df['dE00']

# create a figure and an axis
fig, ax = plt.subplots()

# plot the data inside inside the axis
ax.plot(x,y1, label='Object A')
ax.plot(x,y2, label='Object B')

# display the legend
ax.legend(loc='best',fontsize=12, title='legend title')

# by default the 'loc' value is 'best', but you can select another place: ['upper right', 'lower left', 'upper center', etc.]
# see the documentation for the full list of positions

# show the figure
plt.show()
```
